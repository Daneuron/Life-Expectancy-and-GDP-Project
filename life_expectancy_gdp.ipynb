```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
df = pd.read_csv("all_data.csv")
df.head()
```

```python
#Distribution of GDP in all 6 nations (violin plot)
sns.boxplot(x="Country", y="GDP", data=df, palette="winter")
plt.xticks(rotation=60)
plt.xlabel("Country")
plt.ylabel("GDP 1e13 USD")
plt.title("GDP by Country (2000-2015)")
plt.show()
plt.clf()
```

```python
#Distributionn of Life expectancy for all 6 nations
sns.boxplot(x="Country", y="Life expectancy at birth (years)", data=df, palette="winter")
plt.xticks(rotation=60)
plt.xlabel("Country")
plt.ylabel("Age")
plt.title("Life expectancy at birth by Country (2000-2015)")
plt.show()
plt.clf()
```

```python
#Mapping of Life expectancy vs GDP for all 6 nations
sns.scatterplot(df["Life expectancy at birth (years)"], df["GDP"], hue=df["Country"], palette="bright")
plt.xlabel("Age")
plt.ylabel("GDP 1e13 USD")
plt.title("GDP Vs Life Expectancy by Country")
plt.show()
plt.clf()
```

```python
#Graph to map GDP by Country
plt.plot(df["Year"][df["Country"] == "Chile"], df["GDP"][df["Country"] == "Chile"], label="Chile", color = "green")
plt.plot(df["Year"][df["Country"] == "Mexico"], df["GDP"][df["Country"] == "Mexico"], label="Mexico", color = "blue")
plt.plot(df["Year"][df["Country"] == "Germany"], df["GDP"][df["Country"] == "Germany"], label="Germany", color = "red")
plt.plot(df["Year"][df["Country"] == "United States of America"], df["GDP"][df["Country"] == "United States of America"], label="USA", color = "pink")
plt.plot(df["Year"][df["Country"] == "China"], df["GDP"][df["Country"] == "China"], label="China", color = "orange")
plt.plot(df["Year"][df["Country"] == "Zimbabwe"], df["GDP"][df["Country"] == "Zimbabwe"], label="Zimbabwe", color = "brown")
plt.legend()
plt.xlabel("Year")
plt.ylabel("GDP 1e13 USD")
plt.title("GDP over the years by Country")
plt.show()
plt.clf()
```

```python
#Graph to map Life expectancy by Country
plt.plot(df["Year"][df["Country"] == "Chile"], df["Life expectancy at birth (years)"][df["Country"] == "Chile"], label="Chile", color = "green")
plt.plot(df["Year"][df["Country"] == "Mexico"], df["Life expectancy at birth (years)"][df["Country"] == "Mexico"], label="Mexico", color = "blue")
plt.plot(df["Year"][df["Country"] == "Germany"], df["Life expectancy at birth (years)"][df["Country"] == "Germany"], label="Germany", color = "red")
plt.plot(df["Year"][df["Country"] == "United States of America"], df["Life expectancy at birth (years)"][df["Country"] == "United States of America"], label="USA", color = "pink")
plt.plot(df["Year"][df["Country"] == "China"], df["Life expectancy at birth (years)"][df["Country"] == "China"], label="China", color = "orange")
plt.plot(df["Year"][df["Country"] == "Zimbabwe"], df["Life expectancy at birth (years)"][df["Country"] == "Zimbabwe"], label="Zimbabwe", color = "brown")
plt.legend()
plt.xlabel("Year")
plt.ylabel("Life expectancy")
plt.title("Life expectancy over the years by Country")
plt.show()
plt.clf()
```
